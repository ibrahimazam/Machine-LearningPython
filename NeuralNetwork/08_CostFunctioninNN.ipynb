#### The Cost Function:

<span style="color:Turquoise;">

- It's a function that determines how well a Machine Learning model performs for a given set of data. 
- A cost function is a mechanism utilized in supervised machine learning, the cost function returns the error between predicted outcomes compared with the actual outcomes. The aim of supervised machine learning is to minimize the overall cost, thus optimizing the correlation of the model to the system that it is attempting to represent. 

<span style="color:Orange;">

#### How?

<span style="color:Turquoise;">

- The Training Data Set is used to train an Algorithm in order to produce a trained Model. The Testing Data set is then used to validate the trained Model (which has been split into 2 parts:  80% and 20% for the Training and Testing data sets respectively).
- In Supervised Learning, the Label is the output of the trained model; it is the prediction that the Model makes based on input values. For example, in a binary classification model that decides whether to approve an application for a bank loan, the label would be “Yes/No”.

<span style="color:Orange;">

#### True, False, Positive, Negative

<span style="color:Turquoise;">

- Accuracy (all correct / all) = TP + TN / TP + TN + FP + FN
- Misclassification (all incorrect / all) = FP + FN / TP + TN + FP + FN
- Precision (true positives / predicted positives) = TP / TP + FP
- Sensitivity aka Recall (true positives / all actual positives) = TP / TP + FN
- Specificity (true negatives / all actual negatives) =TN / TN + FP
